```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([1, 2, 3, 4, 5])
y = np.array([10, 40, 50, 78, 83])
epochs = 100
learning_rate = 0.01

plt.plot(X,y, "o",)
plt.show()
```

```python
#import numpy as np

class GradientDescent:
    
    def __init__(self, learning_rate, epochs):
        self.m = 0.0  # Initial value of slope
        self.c = 0.0  # Initial value of intercept
        self.lr = learning_rate
        self.epochs = epochs
        self.lf = None
    
    def fit(self, X, y):
        self.lf = np.zeros(self.epochs)
        self.ValuesOfC=np.zeros(self.epochs)
        for i in range(self.epochs):
            loss_slope_m=sum((self.m*X+self.c-y)*X)/2*len(X)
            loss_slope_c=sum((self.m*X+self.c-y))/2*len(X)
            
            self.m-=self.lr*loss_slope_m
            self.c-=self.lr*loss_slope_c
            self.ValuesOfC[i]=self.c
            self.lf[i] = self.loss_function(X, y)
        plt.plot(self.ValuesOfC,self.lf)
        plt.show()
        print(f"Value of m is {self.m} and value of c is {self.c}")
        
    def loss_function(self, X, y):
        y_pred = self.m * X + self.c
        loss = sum((y_pred-y)**2)/2*len(X)
        return loss
    
    def predict(self, X):
        return self.m * X + self.c
        
    def plot_predictions(self, X, y):
        predictions = self.predict(X)
        print(predictions)
        plt.plot(X, predictions,color='black', label='Optimized Values')
        plt.scatter(X, y, color='red', label='Original Data')
        plt.title('X vs Y')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.legend()
        plt.show()
    
```

```python
gd = GradientDescent(learning_rate, epochs)
gd.fit(X, y)
gd.plot_predictions(X,y)
```
